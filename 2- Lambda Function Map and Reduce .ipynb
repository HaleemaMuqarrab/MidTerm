# Lambda Function 

```python
square = lambda x: x*x
square(6)
```

```python
a = lambda x,y : x+y #short callback function
r = a(2,3)
print(r)
```

```python
multiply= lambda x,y:x*y
```

```python
multiply(405,60)
```

```python
power=lambda x,y :x**y 
```

#reduce fuunction and lambda expression return a single value 

```python
from functools import reduce 
```

```python
nums=[10,20,22,25,29,30]
```

```python
sumup=reduce(lambda x,y:x+y,nums)
```

```python
sumup
```

```python
# reduce function with aggegate function 
```

```python
maxvalue=reduce(lambda x,y:max(x,y), nums)
```

```python
maxvalue
```

```python
maxvalue=reduce(lambda x,y:min(x,y), nums)
```

```python
maxvalue
```
